```python
# Читання тексту з файлу
file_path = "/content/pisnya_short.txt"  # Шлях до вашого файлу

with open(file_path, "r", encoding="utf-8") as file:
    ukrainian_text = file.read().strip()  # Зчитування тексту та видалення зайвих пробілів

#print("Текст із файлу:", ukrainian_text)
```

```python
from transformers import pipeline
```

```python
import re

# Розбиття тексту на речення за допомогою регулярних виразів
sentences = re.split(r'(?<=[.!?])\s+', ukrainian_text)

# Ініціалізація перекладача
translator = pipeline("translation_uk_to_en", model="Helsinki-NLP/opus-mt-uk-en")

# Переклад кожного речення
translated_sentences = [translator(sentence, max_length=512)[0]['translation_text'] for sentence in sentences]

# Об'єднання перекладених речень
translated_text = " ".join(translated_sentences)

# Виведення результату
print("Перекладений текст:\n", translated_text)
```

```python
# Шлях до файлу, куди буде збережено переклад
output_file_path = "translated_text.txt"

# Запис тексту у файл
with open(output_file_path, "w", encoding="utf-8") as file:
    file.write(translated_text)

print(f"Переклад збережено в файл: {output_file_path}")
```

```python
# Переклад тексту (Translation)
translator = pipeline("translation_en_to_uk", model="Helsinki-NLP/opus-mt-en-uk")
translated_text = translator("Hello, how are you?")[0]['translation_text']
print("Переклад тексту:", translated_text)
```

```python
# Аналіз настроїв (Sentiment Analysis)
classifier = pipeline("sentiment-analysis", model="cardiffnlp/twitter-xlm-roberta-base-sentiment")
result = classifier("Коли температура на дворі дорівнює нулю і моросить дощик, це найгірша погода для мене.")[0]
print(f"Аналіз настроїв: {result['label']} з ймовірністю {result['score']:.2f}")
```

```python
# Питання-відповідь (QA)
qa_pipeline = pipeline("question-answering", model="deepset/xlm-roberta-base-squad2")
context = "Андрій Шевченко увійшов до складу збірної України по футболу на Євро 2012, як капітан команди. І в першому матчі чемпіонату забив два м'ячі головою у ворота збірної Швеції."
question = "Хто такий Андрій Шевченко?"
qa_result = qa_pipeline(question=question, context=context)
print(f"Відповідь на запитання: {qa_result['answer']}")
```

Для покращення потрібна редактура з урахуванням культурного контексту та метафоричності тексту. Насправді, переклад не врахував, що Мавка та Лукаш є головними героями, а Лукаш – це ім’я.
